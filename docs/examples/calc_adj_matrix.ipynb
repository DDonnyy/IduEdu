```python
import geopandas as gpd

# Read a GeoDataFrame, for example this one
data = gpd.read_parquet('data/spb_buildings.parquet')

# Create a polygon that encircles all the buildings in the dataset to download the graph
polygon = data.to_crs(4326).union_all().convex_hull.buffer(0.001)
```

```python
from iduedu import get_intermodal_graph, get_drive_graph, get_walk_graph

# Download the intermodal graph
G_intermodal = get_intermodal_graph(territory=polygon)

# Uncomment the following lines to download walking or driving graphs for the same area
# G_walk = get_walk_graph(territory=polygon)
# G_drive = get_drive_graph(territory=polygon)
```

```python
import numpy as np
from iduedu import od_matrix

# Compute the OD matrix between objects in the dataset using the intermodal UrbanGraph.
# If objects are not projected into the graph, the calculation uses their nearest graph nodes.
matrix = od_matrix(
    G_intermodal,
    gdf_origins=data,
    gdf_destinations=data,
    weight="time_min",
    dtype=np.float32,
)

# Uncomment the following lines to compute OD matrices using driving or walking graphs with different weights.
# matrix = od_matrix(G_drive, gdf_origins=data, gdf_destinations=data, weight="time_min", dtype=np.float32)
# matrix = od_matrix(G_walk, gdf_origins=data, gdf_destinations=data, weight="length_meter", dtype=np.float32)
matrix
```
